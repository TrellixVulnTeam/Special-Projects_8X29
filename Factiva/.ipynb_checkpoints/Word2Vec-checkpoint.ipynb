```python
import re
import nltk.data
nltk.download('punkt')
tokenizer = nltk.data.load('nltk:tokenizers/punkt/english.pickle')
```

```python
import nltk
from nltk.corpus import stopwords
set(stopwords.words('spanish'))
```

```python
keyword = ''
f = open(str(keyword) + ".txt", "r+", encoding = 'utf8')
q = f.read()
l=word_tokenize(q)
```

```python
stop_words = set(stopwords.words('spanish'))
word_tok = word_tokenize(q)
filtered_sentence = [w for w in word_tok if not w in stop_words]
sent = ' '.join(filtered_sentence)
```

```python
def sentence_to_wordlist(sentence, remove_stopwords=False):
    # 1. Remove non-letters
    sentence_text = re.sub(r'[^\w\s]','', sentence)
    # 2. Convert words to lower case and split them
    words = sentence_text.lower().split()
    # 3. Return a list of words
    return(words)
```

```python
def oped_to_sentences(oped, tokenizer, remove_stopwords=False):
    try:
        # 1. Use the NLTK tokenizer to split the text into sentences
        raw_sentences = tokenizer.tokenize(oped.strip())
        # 2. Loop over each sentence
        sentences = []

        for raw_sentence in raw_sentences:
            # If a sentence is empty, skip it
            if len(raw_sentence) > 0:
                # Otherwise, call sentence_to_wordlist to get a list of words
                sentences.append(sentence_to_wordlist(raw_sentence))
        # 3. Return the list of sentences (each sentence is a list of words, so this returns a list of lists)
        len(sentences)
        return sentences
    except:
        print('nope')
```

```python
sent_list = oped_to_sentences(sent, tokenizer)
```

```python
import gensim
from gensim.models import Word2Vec
from gensim.models import word2vec
from gensim.models import Phrases
import logging
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
num_features = 100    # Word vector dimensionalty
min_word_count = 15   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size                                                                                    
downsampling = 1e-3 
```

```python
model = gensim.models.Word2Vec(sent_list, workers=num_workers, size=num_features, min_count = min_word_count, window = context, sample = downsampling)
```

```python
model.save(keyword)
```

```python
print(model)
```

```python

f = open(str(keyword) + '.txt','w', encoding='utf8')
f.write(str(model.most_similar(positive=keyword', topn=4000)))
f.close()
```
